```python


def get_files_by_type(file_list):
    """Separate files into categories based on extension."""
    pdf_files = []
    csv_files = []
    json_files = []

    for file in file_list:
        if file.lower().endswith('.pdf'):
            pdf_files.append(file)
        elif file.lower().endswith('.csv'):
            csv_files.append(file)
        elif file.lower().endswith('.json'):
            json_files.append(file)
    
    return pdf_files, csv_files, json_files


def display_files(pdf_files, csv_files, json_files):
    """Display categorized files."""
    print("PDF Files:")
    for f in pdf_files:
        print("  -", f)

    print("\nCSV Files:")
    for f in csv_files:
        print("  -", f)

    print("\nJSON Files:")
    for f in json_files:
        print("  -", f)


def main():
    # Example file list (you can also use os.listdir(path) to get actual files)
    files = [
        "report.pdf", "data.csv", "notes.txt", "info.json", 
        "summary.PDF", "sales_data.CSV", "config.JSON", "image.png",'hkp.pdf'
    ]

    pdf_files, csv_files, json_files = get_files_by_type(files)
    


if __name__ == "__main__":
    main()
```
