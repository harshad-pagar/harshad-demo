```python
#prime number
n=int(input('enter a no'))
if n>1:
    for i in range(2,n):
        if n%i==0:
            print('not prime')
            break
    else:
        print('prime')
else:
    print("please enter a no greater than 1 ")
            
```

```python
#armstrong
num=int(input('enter a no:'))
original=num
digit=num
digits = len(str(num))
sum=0
while num > 0:
    digit = num % 10
    sum=sum + (digit ** digits)
    num=num //10
if sum == original:
    print('armstrong')
else:
    print('not')
```
