```python
lambda function:-
A lambda function in Python is a small, anonymous (nameless) function used for short tasks.
It is written in one line using the lambda keyword, 
and is mostly used when we don't need to reuse the function.
```

```python
def app(p,value):
    return 6 + p(value)
print(app(lambda x:x**2,2))
```

```python
def decor(func):
    def inner():
        print('*'*20)
        func()
        print('*'*20)
    return inner
        
@decor
def show():
    print("HARSHAD")

show()
```
