```python
#FETRCH NO. FROM FILE
def fetch_no(ft):
    num=[]
    for i in ft:
        if i.isnumeric():
            num.append(i)
    return num
def fetch_ch(ft):
    char=[]
    for i in ft:
        if i.isalpha():
            char.append(i)
    return char

def fetch_word(ft):
    word=[]
    for i in ft:
        if i.isword():
            word.append(i)
    return word
def read_content(fn):
    fp=open(fn)
    data=fp.read()
    print(data)
    return data
def main():
    filename="test.txt"
    fd=read_content(filename)
    n=fetch_no(fd)
    n1=fetch_ch(fd)
    n2=fetch_word(fd)
    print(n)
    print(n1)
    print(n2)
    

if __name__=='__main__':
    main()
```
