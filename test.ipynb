```python
a=int(input("enter a no."))
factorial=1
if a<0:
    print("factorial not for negative no.")
elif a==0:
    print("factorial of 0 is 1")
else:
    for i in range(1,a+1):
        factorial=factorial * i   #1*1=1,1*2=2,2*3=6,6*4=12 (main logic)
    print("factorial  of ",a,"is",factorial)
   
```

```python
def vowels_count(paragraph):
    vowels="aeiouAEIOU"
    count=0
    for char in paragraph:
        if char in vowels:
            count+=1
    return count

paragraph = """Madam Arora teaches Malayalam at noon. She enjoys reading books and listening to music."""
print("Total count:", vowels_count(paragraph))
    
```
